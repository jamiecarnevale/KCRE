```python
df1 = kchd.head(2)
```

```python
for index, row in df1.iterrows():
    
    result = geocoder.google([row['lat'], row['long']], method='reverse', key=api_key)
    print(result.raw)
```

```python
zips =[]
hoods = []


for index, row in df1.iterrows():
    result = geocoder.google([row['lat'], row['long']], method='reverse', key=api_key)
    zips.append(dict(result.raw)['postal_code']['long_name'])
    hoods.append(result.raw['address_components'][2]['long_name'])
df1['zips'] = zips
df1['hoods'] = hoods
```

```python
zips = []
hoods = []

for index, row in df1.iterrows():
    result = geocoder.google([row['lat'], row['long']], method='reverse', key=api_key)

    # Error handling for postal_code
    try:
        zip_code = dict(result.raw)['postal_code']['long_name']
    except KeyError:
        zip_code = None
    zips.append(zip_code)

    # Error handling for neighborhood
    try:
        neighborhood = result.raw['address_components'][2]['long_name']
    except (KeyError, IndexError):
        neighborhood = None
    hoods.append(neighborhood)

df1['zips'] = zips
df1['hoods'] = hoods
```

```python
hoods = []

for i in tqdm(range(10000)):
    for index, row in dfx.iterrows():
        result = geocoder.google([row['lat'], row['long']], method='reverse', key=api_key)

    # Error handling for neighborhood
        try:
            neighborhood = result.raw['address_components'][2]['long_name']
        except (KeyError, IndexError):
            neighborhood = None
        hoods.append(neighborhood)

dfx['hoods'] = hoods
```

```python
hoods = []

# Using the progress bar with the iteration over the dataframe rows
for index, row in tqdm(dfx.iterrows(), total=dfx.shape[0]):
    result = geocoder.google([row['lat'], row['long']], method='reverse', key=api_key)

    # Error handling for neighborhood
    try:
        neighborhood = result.raw['address_components'][2]['long_name']
    except (KeyError, IndexError):
        neighborhood = None
    hoods.append(neighborhood)

dfx['hoods'] = hoods
```
